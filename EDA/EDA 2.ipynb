## 3. Box Plot (Numerical - Categorical)

# 4. Distplot (Numerical - Categorical)

```python
sns.distplot(x=titanic[titanic['Survived']==0]['Age'],hist=False)
sns.distplot(x=titanic[titanic['Survived']==1]['Age'],hist=False)
```

# 5. HeatMap (Categorical - Categorical)

```python
pd.crosstab(titanic['Pclass'],titanic['Survived'])
```

```python
sns.heatmap(pd.crosstab(titanic['Pclass'],titanic['Survived']),annot=True,fmt=".1f")
```

```python
# Group by 'sex' column and calculate the mean age for each gender
mean_age_by_sex = titanic.groupby('Pclass')['Survived'].mean()
print(mean_age_by_sex)
```

```python
(titanic.groupby('Pclass')['Survived'].mean())*100
```

```python
(titanic.groupby(['Pclass','Embarked'])['Survived'].mean())*100
```
